Checking metrice are valid (non-negative).

```python
List=['Impressions','Clicks','Conversion_Rate','ROI']
for x in List:
    x=(df[x]>=0).all()
    print(x)
```

removing noice from columns

```python
df['Acquisition_Cost']=df['Acquisition_Cost'].str.strip('$')
```

```python
df['Acquisition_Cost']=df['Acquisition_Cost'].str.replace(',','')
```

```python
df['Acquisition_Cost']=df['Acquisition_Cost'].astype('float64')
```

Saving the csv-file

```python
df.to_csv('Marketing_Campaign_Performance_Dataset2.csv',index=False)
```

```python
df.columns
```

Calculating Key metrics 

```python
df['cost']=df['Clicks']*df['Acquisition_Cost']
df['CTR_(Click-Through Rate)']=df['Clicks']/df['Impressions']
df['CPC (Cost per Click)']=df['cost']/df['Clicks']
df['Conversions']=df['Conversion_Rate']*df['Clicks']
df['CPA (Cost per Acquisition)']=df['cost']/df['Conversions']
#→ Revenue = Cost × (1 + (ROI / 100))
df['Revenue']=df['cost']*(1+(df['ROI']/100))
#ROI (%) = ((Revenue - Cost) / Cost) × 100
df['ROI (%)']=(df['Revenue']-df['cost'])*100
```
